```python
import accelerate
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load model directly
tokenizer = AutoTokenizer.from_pretrained("01-ai/Yi-6B")
model = AutoModelForCausalLM.from_pretrained("01-ai/Yi-6B")
```

**Test with the default model**

```python
# default model
import torch

#device = "cuda:0" if torch.cuda.is_available() else "cpu"

while True:
    # Prompt content
    prompt = input("Enter your prompt (or 'exit' to quit): ")

    # Check if the user wants to exit
    if prompt.lower() == 'exit':
        break

    # Tokenize the input prompt
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate text
    output_ids = model.generate(input_ids, max_length=80, num_beams=5, no_repeat_ngram_size=2)

    # Decode and print the generated text
    generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Ensure the output is a complete sentence
    if '.' in generated_text:
        generated_text = generated_text.split('.')[0] + '.'

    print("Generated text:", generated_text)
```
